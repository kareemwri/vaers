```python
vaers=pd.read_csv(r'D:\Documents\Projects\VAERS\vaers.csv')
```

```python
vaers['RECVDATE']=vaers.RECVDATE.fillna('01/01/1900')
vaers['DATEDIED']=vaers.DATEDIED.fillna('01/01/1900')
vaers['VAX_DATE']=vaers.VAX_DATE.fillna('01/01/1900')
vaers['ONSET_DATE']=vaers.ONSET_DATE.fillna('01/01/1900')
vaers['STATE']=vaers.STATE.fillna('UKNOWN')
vaers['SEX']=vaers.SEX.fillna('U')
vaers['DIED']=vaers.DIED.fillna('UKNOWN')
vaers['VAX_TYPE']=vaers.VAX_TYPE.fillna('UKNOWN')
vaers['VAX_MANU']=vaers.VAX_MANU.fillna('UKNOWN')
vaers['NUMDAYS']=vaers.NUMDAYS.fillna('0')
vaers['AGE_YRS']=vaers.AGE_YRS.fillna('100')
```

```python
vaers.info()
```

```python
vaers['VAERS_ID']=vaers.VAERS_ID.astype('int32')
vaers['RECVDATE']=pd.to_datetime(vaers['RECVDATE'])
vaers['DATEDIED']=pd.to_datetime(vaers['DATEDIED'])
vaers['VAX_DATE']=pd.to_datetime(vaers['VAX_DATE'])
vaers['ONSET_DATE']=pd.to_datetime(vaers['ONSET_DATE'])
vaers['NUMDAYS']=vaers.NUMDAYS.astype('float64')
vaers['AGE_YRS']=vaers.AGE_YRS.astype('float64')
```
